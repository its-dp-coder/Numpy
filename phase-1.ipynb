### List vs numpy Array

```python
list=[1,2,3,4]
print(f" List in Python {list*4}")

np_array=np.array([6,7,8,9])  #Element wise multiplication
print(f"Python Array Multiplication {np_array*2}")

start=t.time()
list=[i*2 for i in range(1000000)]
print("\n List operation time: ",t.time()-start)


start=t.time()
np_array=np.arange(1000000) * 2
print("\n Numpy  operation time: ",t.time()-start)
```
